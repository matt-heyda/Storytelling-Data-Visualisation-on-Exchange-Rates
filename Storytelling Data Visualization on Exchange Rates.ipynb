# Storytelling Data Visualisation on Exchange Rates: How the Euro-Dollar Rate has Changed During the Coronavirus Pandemic

## Introduction

The dataset used for this project describes Euro daily exchange rates between 1999 and 2021. The euro (symbolised with €) is the official currency in most of the countries of the European Union. Our focus will be on the exchange rate between the euro and the American dollar and how this has changed during the coronavirus pandemic.

The data was made available by Daria Chemkaeva on [Kaggle](https://www.kaggle.com/datasets/lsind18/euro-exchange-daily-rates-19992020), and the data source is the European Central Bank. Note that this data is updated regularly, however, the set used here was downloaded on January 2021.

## Introducing the Dataset

```python
import pandas as pd

exchange_rates = pd.read_csv("euro-daily-hist_1999_2020.csv")

exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

The dataset contains 41 columns, each describing a different currency, and 5699 rows, ranging from 01/04/1999 to 01/08/2021. Most of the data is of the 'object' dtype (i.e. strings) and some columns contain many null values. Thus, we have some data cleaning to do:

## Data Cleaning

We will do the following:

* Rename the `[US dollar ]` and `Period\Unit:` columns to something easier to type - `US_dollar` and `Time`.
* Change the `Time` column to a `datetime` data type.
* Sort the values by `Time` in ascending order.
* Reset the index (and drop the initial index).

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

Isolating the `Time` and the `US_dollar` columns:

```python
euro_to_dollar = exchange_rates[["Time", "US_dollar"]].copy()

euro_to_dollar["US_dollar"].value_counts()
```

The value counts shows us that we have 62 rows that contain the `-` character. We will drop these rows:

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar["US_dollar"] != "-"]
```

```python
## Converting US_dollar to `float` ##

euro_to_dollar["US_dollar"] = euro_to_dollar["US_dollar"].astype(float)
```

## Moving Average (Rolling Mean)

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar["Time"],
         euro_to_dollar["US_dollar"])
plt.show()
```

The line graph above shows us the visual representation of the daily variation in the exchange rate. However, we want to focus on longer-term trends and, therefore, will use a 30-day moving average to visualise our data.

```python
## Creating new 'rolling_mean' column ##

euro_to_dollar["rolling_mean"] = euro_to_dollar["US_dollar"].rolling(30).mean()
euro_to_dollar
```

```python
## Simple line graph with rolling mean ##

plt.plot(euro_to_dollar["Time"], euro_to_dollar["rolling_mean"])
plt.title("Exchange Rate: Euro to US Dollar (30-day Moving Avg.)", weight="bold")
plt.xlabel("Year")
plt.ylabel("US Dollars")
plt.show()
```

## Storytelling: How the Euro-Dollar Rate Has Changed During the Coronavirus Pandemic

We can show the 2020 data and the 2016-2019 data as a baseline. We can use a line plot.
